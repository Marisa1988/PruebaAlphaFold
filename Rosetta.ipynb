```python
import requests

# Base URL for the GitHub API
base_url = "https://api.github.com/repos/RosettaCommons/rosetta"

# Function to fetch and display repository data
def fetch_repo_data():
    # Fetch repository details
    repo_response = requests.get(base_url)
    if repo_response.status_code == 200:
        repo_info = repo_response.json()
        print("Repository Information:")
        print(f"Name: {repo_info['name']}")
        print(f"Description: {repo_info['description']}")
        print(f"Owner: {repo_info['owner']['login']}")
        print(f"Stars: {repo_info['stargazers_count']}")
        print(f"Forks: {repo_info['forks_count']}")
        print(f"Open Issues: {repo_info['open_issues_count']}")
        print(f"URL: {repo_info['html_url']}")
    else:
        print(f"Failed to retrieve repository details: {repo_response.status_code}")

    # Fetch branches
    branches_response = requests.get(f"{base_url}/branches")
    if branches_response.status_code == 200:
        branches = branches_response.json()
        print("\nBranches:")
        for branch in branches:
            print(f"- {branch['name']}")
    else:
        print(f"Failed to retrieve branches: {branches_response.status_code}")

    # Fetch contributors
    contributors_response = requests.get(f"{base_url}/contributors")
    if contributors_response.status_code == 200:
        contributors = contributors_response.json()
        print("\nContributors:")
        for contributor in contributors:
            print(f"- {contributor['login']}")
    else:
        print(f"Failed to retrieve contributors: {contributors_response.status_code}")

    # Fetch issues
    issues_response = requests.get(f"{base_url}/issues")
    if issues_response.status_code == 200:
        issues = issues_response.json()
        print("\nOpen Issues:")
        for issue in issues:
            print(f"- {issue['title']}")
    else:
        print(f"Failed to retrieve issues: {issues_response.status_code}")

# Run the function to fetch and display data
fetch_repo_data()
```

```python
import requests
import base64

# GitHub API URL for the README file of the PyRosetta.notebooks repository
url = "https://api.github.com/repos/RosettaCommons/PyRosetta.notebooks/readme"

# Make a GET request to the GitHub API
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    readme_info = response.json()
    
    # Decode the content from base64
    readme_content = base64.b64decode(readme_info['content']).decode('utf-8')
    
    # Print the README content
    print("Instructions on How to Get Started, Basic Operations, and Accessing Documentation:\n")
    print(readme_content)
else:
    print(f"Failed to retrieve README file: {response.status_code}")
```
